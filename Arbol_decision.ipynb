```python
#Importamos las librerias 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
```

```python
#Cargamos el dataset
#mostramos algunos datos
data = pd.read_csv("winequality-red.csv")
data.head()
```

```python
#cantidad de registros y columnas
data.shape
```

```python
data.describe()
```

```python
#validamos si se presentan valores nulos
data.isna().sum()
```

```python
#histograma donde vemos las cantidades segun la calificion de vino
plt.hist(data.quality)
```

```python
#diagrama de dispersion donde tome las variables acidez fija y  acidez volatil
sns.scatterplot(x = "fixed acidity", y = "volatile acidity", hue = "quality", data = data)
```

```python
#dividimos el dataset
predictors_col = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", 
             "total sulfur dioxide", "density", "pH", "sulphates", "alcohol"]
target_col = ['quality']
```

```python
predictors = data[predictors_col]
target = data[target_col]
```

```python
#se valida si quedo bien dividido el dataset
predictors
```

```python
target
```

```python
#correlacion entre variables predictoras
corr = predictors.corr()
corr
```

```python
#mapa de calor (correlacion)
sns.heatmap(corr)
plt.show()
```

```python
#se realiza la division del dataset donde indicamos el 25% apara test y el 75% para train 
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.25, random_state=13)
```

```python
#creamos arbol de decision
tree = DecisionTreeClassifier()
```

```python
#entreamos nuestro DATA SET
arbol = tree.fit(X_train, y_train)
```

```python
#mostramos la grafica del arbol creado
plot_tree(arbol)
```

```python
predicciones = arbol.predict(X_test)
```

```python
predicciones
```

```python
# Generamos una matroz de confusion
pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), predicciones, rownames=['Actual'], colnames=['Predicciones'])
```

```python
#medida que permite validar que tan cerca estuvieron las predicciones 
accuracy = accuracy_score(y_test,predicciones)
accuracy
```
