```python
# importar datos
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
#Librería para Regresión Lineal
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
#cargar archivo de vehiculos
Data = pd.read_csv('./Car details v3.csv')
```

```python
# Mostramos informacion de archivo
Data.head(15)
```

```python
#Descripcinn de los datos
Data.describe()
```

```python
# se realzia la depuracion reemplazando por ceros los valores nuelos y sacando el minimo para los valores nuelos en seats
# decidi eliminar los registros que estaban en 0 debido a que son importantes a la hora de comprar un vehiculo y los elimino del dataset
Data['mileage'].fillna(0, inplace = True)
Data['engine'].fillna(0, inplace = True)
Data['max_power'].fillna(0, inplace = True)
Data['torque'].fillna(0, inplace = True)
Data['seats'].fillna(Data['seats'].min(), inplace = True)
```

```python
#buscamos valores vacios
Data.isnull().sum()
```

```python
#Elimino los registros en 0
Data = Data.drop(Data[Data.mileage == 0].index)
Data = Data.drop(Data[Data.engine == 0].index)
Data = Data.drop(Data[Data.max_power == 0].index)
Data = Data.drop(Data[Data.torque == 0].index)
```

```python
#Demostracion de valores en 0 eliminados
Data.head(15)
```

```python
#tomo los campos año y kilomtreaje recorrido como variables indpendientes
x=np.array(Data[['year','km_driven']])
y=Data[['selling_price']].values
```

```python
modelo=LinearRegression()
```

```python
modelo.fit(x,y)
```

```python
print(f"La ecuacion de la recta es: y ={modelo.intercept_}+{modelo.coef_[0]}*x")
```

```python
y_pred = modelo.predict(x)
mse = mean_squared_error(y, y_pred)
rmse=np.sqrt(mse)
mae= mean_absolute_error(y, y_pred)
r2= r2_score(y, y_pred)
```

```python
print(f" Error cuadratico medio (MSE) : {mse}")
```

```python
print(f" Raiz del error cuadratico medio (RMSE) : {rmse}")
```

```python
print(f" Error absoluto medio (MAE) : {mae}")
```

```python
print(f" El coeficiente de determinacion (R2) : {r2}")
```
